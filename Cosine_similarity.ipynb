```python
import math
import os

def calc_dot_product(doc, query):
    product_val=0
    for word in doc :
        if word in query:
            product_val+=doc[word]*query[word]
    return product_val

def calc_cosine_similarity(doc, query):
    dot_product=calc_dot_product(doc, query)
    #return math.degrees( math.acos(dot_product/(math.sqrt(calc_dot_product(doc, doc)*calc_dot_product(query, query)))))
    return (dot_product/(math.sqrt(calc_dot_product(doc, doc)*calc_dot_product(query, query))))

def question1():

    #read query file
    dot_product_total=[]
    cosine_similarity_total=[]
    path=r'./queries'
    for filename in os.listdir(path):
        path_q=str(os.path.join(path, filename))
        query=open(path_q, 'r')
        #create list of words in query
        q=[]
        for lines in query:
            l=lines.split('\n')
            q.append(l[0])
        #print(q)

        #create hash table for frequency of words in query
        dictionary_query = {}
        for word in q:
            if word in dictionary_query:
                dictionary_query[word]=dictionary_query[word]+1
            else:
                dictionary_query[word]=1
        #print(dictionary_query)

        #list to store data for all docs corresponding to current query
        dot_product_individual_query=[]
        cosine_similarity_individual_query=[]

        #iterate through all docs:
        filenames=[]
        path_doc = r'./docs'
        for filename1 in os.listdir(path_doc):
            data = open(os.path.join(path_doc, filename1), 'r')
            filenames.append(str(filename1))
            # create list of words from doc
            d = []
            for lines in data:
                l = lines.split('\n')
                d.append(l[0])
            #print(d)

            # create hash table for frequency of words
            dictionary_doc = {}
            for word in d:
                if word in dictionary_doc:
                    dictionary_doc[word] = dictionary_doc[word] + 1
                else:
                    dictionary_doc[word] = 1
            #print(dictionary_doc)

            dot_product_result=calc_dot_product(dictionary_doc, dictionary_query)
            #print(dot_product_result)
            dot_product_individual_query.append(dot_product_result)
            #print(dot_product_individual_query)

            cosine_similarity_result=calc_cosine_similarity(dictionary_doc, dictionary_query)
            #print(cosine_similarity_result)
            cosine_similarity_individual_query.append(cosine_similarity_result)


        dot_product_total.append(dot_product_individual_query)
        #print(dot_product_individual_query)
        #print(dot_product_total)

        cosine_similarity_total.append(cosine_similarity_individual_query)
        #print(cosine_similarity_individual_query)
        #print(cosine_similarity_total)


    #print("dot product total:\n")
    #print(dot_product_total)
    #print()
    #print("cosine similarity total:\n")
    #print(cosine_similarity_total)
    #print()

    outpath = r'./'
    #get top 10 similar docs for each query based on DOT PRODUCT - getting the top 10 after descending order sort(more frequencies together- more similarity)
    #for each query:
    #create hash_table with index of doc file and value
    #sort hashtable based on values in descending order
    #keep only 10 values

    dot_out = open(os.path.join(outpath, 'dot_product_output_data.txt'), 'w+')
    dot_out.write('Format : (file,  dot product value)\n')
    print('Format : (file,  dot product value)\n')
    #dot_out.close()
    list_of_dot_product_dictionaries_for_each_query=[]
    for i in range(0, len(dot_product_total)):
        dot_product_dictionary = {}
        for j in range(0, len(dot_product_total[0])):
            dot_product_dictionary[filenames[j]]=dot_product_total[i][j]

        sorted_dot = sorted(dot_product_dictionary.items(), key=lambda x: x[1], reverse=True)
        list_of_dot_product_dictionaries_for_each_query.append(sorted_dot[:10])

    #dot_out=open(os.path.join(outpath, 'dot_product_output_data.txt'), 'w+')
    for i in range(0, len(list_of_dot_product_dictionaries_for_each_query)):
        dot_out.write('query='+str(i)+':\n')
        print('query='+str(i)+':\n')
        for j in range(0, 10):
            dot_out.write(str(list_of_dot_product_dictionaries_for_each_query[i][j])+'\n')
            print(str(list_of_dot_product_dictionaries_for_each_query[i][j])+'\n')
    dot_out.close()

    print('----------------------------------------------------------------------------------------------------------')

    #print()

    #get top 10 similar docs for each query based on COSINE SIMILARITY - getting the top 10 after descending order sort since cos 0=1 (similar) and cos 90=0(different)
    #for each query:
    #create hash_table with index of doc file and value
    #sort hashtable based on values in descending order
    #keep only 10 values
    cosine_out = open(os.path.join(outpath, 'cosine_similarity_output_data.txt'), 'w+')
    cosine_out.write('Format : (file,  cosine similarity value)\n')
    print('\n\nFormat : (file,  cosine similarity value in cos theta)\n')
    list_of_cosine_similarity_dictionaries_for_each_query=[]
    for i in range(0, len(cosine_similarity_total)):
        cosine_similarity_dictionary = {}
        for j in range(0, len(cosine_similarity_total[0])):
            cosine_similarity_dictionary[filenames[j]]=cosine_similarity_total[i][j]

        sorted_dot_cosine = sorted(cosine_similarity_dictionary.items(), key=lambda x: x[1], reverse=True)
        list_of_cosine_similarity_dictionaries_for_each_query.append(sorted_dot_cosine[:10])
    

    for i in range(0, len(list_of_cosine_similarity_dictionaries_for_each_query)):
        cosine_out.write('query=' + str(i) + ':\n')
        print('query=' + str(i) + ':\n')
        for j in range(0, 10):
            cosine_out.write(str(list_of_cosine_similarity_dictionaries_for_each_query[i][j]) + '\n')
            print(str(list_of_cosine_similarity_dictionaries_for_each_query[i][j]) + '\n')
    cosine_out.close()

    print()



        #print(lines)
        #print()

```
